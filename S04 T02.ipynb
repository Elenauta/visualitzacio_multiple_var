```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
tips = pd.read_csv('tips.csv')
tips.head(3)
```

```python
tips.shape
```

```python
tips.columns
```

```python
tips['day'].unique()
```

```python
tips['day'].nunique()
```

```python
# Create a new column 'tip per person' and get rid of the 'size' one
```

```python
tips['tip_person'] = tips['tip'] / tips['size']
```

```python
tips.drop(['size'], axis=1, inplace=True)
```

```python
tips.describe().round(2)
```

```python
tips.total_bill.mean()
```

```python
x = tips.total_bill
y = tips.tip
colors = 'red'



plt.axis([0, 60, 0, 15])
plt.scatter(x, y, c=colors, alpha=0.4)

plt.title('Total Bill vs Tips', )
plt.ylabel('Tip')
plt.xlabel('Total Bill')
plt.show()
```

```python
plot = tips['sex'].value_counts().plot(kind='barh', title='Tips by Sex', color='g')
plt.show() 
```
